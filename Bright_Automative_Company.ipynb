## 	Descriptive Statistics:
What are the mean, median, and standard deviation of the ages of individuals in the dataset?

```python
# Mean
mean_age = df['Age'].mean()

# Median
median_age = df['Age'].median()

# Standard deviation
std_age = df['Age'].std()

print(f"Mean age: {mean_age}")
print(f"Median age: {median_age}")
print(f"Standard Deviation of age: {std_age}")
```

## 	Data Distribution:
What is the distribution of gender in the dataset? Represent it using a pie chart.

```python
import matplotlib.pyplot as plt

# Count the frequency of each gender
gender_counts = df['Gender'].value_counts()

# Plot pie chart
plt.figure(figsize=(5,5))
plt.pie(gender_counts, labels=gender_counts.index)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

###	Correlation Analysis:
Is there a correlation between age and salary? Provide the correlation coefficient and interpret the result.

```python
# Correlation coefficient between 'Age' and 'Salary'
correlation = df['Age'].corr(df['Salary'])

print(f"Correlation coefficient between Age and Salary: {correlation:.3f}")
```

###Salary Analysis:
####What is the average salary for individuals based on their educational qualifications (Graduate vs. Post Graduate)?


```python
sns.barplot(data=df, x='Education', y='Salary', estimator='mean', ci=None)
plt.title('Average Salary by Education Level')
plt.ylabel('Average Salary')
plt.xlabel('Education Level')
plt.show()
```

###Loan Status:
####What percentage of individuals have taken a personal loan? How does this compare between males and females?


```python
df['Personal_loan_numeric'] = df['Personal_loan'].apply(lambda x: 1 if x == 'Yes' else 0)

total_loan_percentage = (df['Personal_loan_numeric'].sum() / len(df)) * 100
print(f"Percentage of individuals who have taken a personal loan: {total_loan_percentage:.2f}%")

male_loan_percentage = (df[df['Gender'] == 'Male']['Personal_loan_numeric'].sum() / df[df['Gender'] == 'Male'].shape[0]) * 100
print(f"Percentage of males who have taken a personal loan: {male_loan_percentage:.2f}%")

female_loan_percentage = (df[df['Gender'] == 'Female']['Personal_loan_numeric'].sum() / df[df['Gender'] == 'Female'].shape[0]) * 100
print(f"Percentage of females who have taken a personal loan: {female_loan_percentage:.2f}%")

sns.barplot(data=df, x='Gender', y='Personal_loan_numeric', estimator='mean', ci=None)
plt.title('Personal Loan Status by Gender')
plt.ylabel('Proportion of Individuals with Personal Loan')
plt.xlabel('Gender')
plt.show()
```

###Marital Status and Dependents:
####What is the average number of dependents for married individuals versus single individuals?


```python
df['No_of_Dependents'] = pd.to_numeric(df['No_of_Dependents'], errors='coerce')

df.dropna(subset=['No_of_Dependents'], inplace=True)

average_dependents_by_marital_status = df.groupby('Marital_status')['No_of_Dependents'].mean()

print("Average number of dependents by marital status:")
print(average_dependents_by_marital_status)

sns.barplot(data=df, x='Marital_status', y='No_of_Dependents', estimator='mean', ci=None)
plt.title('Average Dependents by Marital Status')
plt.ylabel('Average Number of Dependents')
plt.xlabel('Marital Status')
plt.show()
```
